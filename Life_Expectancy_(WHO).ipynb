```python
# Calculate number of outliers and its percentage in each variable using Tukey's method.

for variable in col_dict.keys():
    q75, q25 = np.percentile(df[variable], [75 ,25])
    iqr = q75 - q25

    min_val = q25 - (iqr*1.5)
    max_val = q75 + (iqr*1.5)
    print("Number of outliers and percentage of it in {} : {} and {}".format(variable,len((np.where((df[variable] > max_val) | (df[variable] < min_val))[0])),
                                                                                      len((np.where((df[variable] > max_val) | (df[variable] < min_val))[0]))*100/2938))
```
